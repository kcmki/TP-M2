```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

# Reading the data

```python

Data = pd.read_excel('Exo1.xlsx')
Data_np = np.array(Data)
Data_np = Data_np[:,1:]
```

# Diagramme de dispersion

```python

plt.scatter(Data_np[:,0], Data_np[:,1])

# Ajouter des labels et un titre
plt.xlabel('Longeur')
plt.ylabel('Largeur')
plt.title('Diagramme de Dispersion de la largeur et longeur des pétales')

# Afficher le diagramme
plt.show()
```

# Histogramme

```python
plt.hist(Data_np[:,0], bins=5, edgecolor='black')  # Vous pouvez ajuster le nombre de bacs (bins) selon vos besoins.

# Ajouter des labels et un titre
plt.xlabel('Longeur des pétales')
plt.ylabel('Fréquence')
plt.title('Histogramme Simple des largeurs des pétales')

# Afficher le diagramme
plt.show()
```

# Boite a moustache

```python
plt.boxplot(Data_np[:,1])

# Ajouter un titre
plt.title('Boîte à Moustache')

# Afficher le graphique
plt.show()
```
