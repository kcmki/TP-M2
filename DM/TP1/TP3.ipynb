```python

import numpy as np

import math

import matplotlib

import pandas as pd
```

# Reading files and setting up parameters

```python
def sepLine(line):
    return line.split(";")[1:]
def getArray(file):
    arr = []
    for line in file:
        line = sepLine(line)
        Line = []
        for x in line:

            x = x.replace(",",".")

            try:
                Line.append(float(x))
            except:
                Line.append(x)
        arr.append(Line)
    return arr


f = "Exo1.csv"
file = open(f,"r")

array = getArray(file)
array = array[1:]

k = 5
```

```python
def min(array):
    mini = array[0]
    for i in range(len(array)):
        if mini > array[i]:
            mini = array[i]
    return mini
def max(array):
    maxi = array[0]
    for i in range(len(array)):
        if array[i] > maxi:
            maxi = array[i]
    return maxi
```

# Discretize

discretisation des données (echanger les valeur situé dans un meme interval par un singe commun)

```python
def intervale(Dataset,k=None):

    if k == None:
        k=1+(3/10)*math.log10(len(Dataset))

    intervale = []
    mins = []
    maxs = []
    for i in range(len(Dataset[0]) - 1):
        mini = min([Dataset[j][i] for j in range(len(Dataset))])
        mins.append(mini)
        maxi = max([Dataset[j][i] for j in range(len(Dataset))])
        maxs.append(maxi)
        intervale.append((maxi - mini)/k)

    return intervale

def discretise(Dataset,intervale):

    DiscData = []
    for i in range(len(Dataset)):
        row = []
        for j in range(len(intervale)):
            if int(((Dataset[i][j] - mins[j])/intervale[j])) == k:
                row.append(int(((Dataset[i][j] - mins[j])/intervale[j]))-1)
            else:
                row.append(int(((Dataset[i][j] - mins[j])/intervale[j])))
            
        DiscData.append(row)

    return DiscData , mins , maxs

intervale = intervale(array,k=k)
DiscData , mins , maxs = discretise(array,intervale)
DiscData
```

# Remplacement par la moyenne des intervales

```python
def getMoyennes(Dataset,DiscData,k=None):
    if k == None:
        k=1+(3/10)*math.log10(len(Dataset))

    sommes = np.zeros((k,len(Dataset[0]) - 1))
    count = np.zeros((k,len(Dataset[0]) - 1))
    moyennes = []

    for i in range(len(Dataset)):
        for j in range(len(Dataset[0]) - 1):
            
            sommes[DiscData[i][j]][j] += Dataset[i][j]
            count[DiscData[i][j]][j] += 1

    return np.divide(sommes,count)

def DiscToMoyenne(DiscData,moyenne):
    moy = np.zeros((len(DiscData),len(DiscData[0])))
    for i in range(len(DiscData)):
        for j in range(len(DiscData[0])):

            moy[i,j] = moyenne[DiscData[i][j]][j]

    return moy  
          
moyenne = getMoyennes(array,DiscData,k=k)
MeanData = DiscToMoyenne(DiscData,moyenne)

MeanData
```

# Normalisation des données

methode min max

```python
def NormalizeMinMax(Dataset,mins,maxs,Nmin,Nmax):
    NormData = np.zeros((len(Dataset),len(Dataset[0])))

    for i in range(len(Dataset)):
        for j in range(len(Dataset[0])):
            NormData[i][j] = (Dataset[i][j] - mins[j])/(maxs[j] - mins[j])*(Nmax - Nmin) + Nmin
    
    return NormData


NormData = NormalizeMinMax(MeanData,mins,maxs,0,1)

NormData
```
